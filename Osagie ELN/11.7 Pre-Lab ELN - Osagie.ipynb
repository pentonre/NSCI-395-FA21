### Including sex differences
I use a for loop to examine the metadata of each mouse experiment containing natural scenes and separate the experiment container IDs by the sex of the mouse.

```python
exps = boc.get_ophys_experiments(stimuli=['natural_scenes'])

male_ec_ids = []
female_ec_ids = []

for i in range(len(exps)):
    exp = exps[i]
    exp_data = boc.get_ophys_experiment_data(exp['id'])
    
    if(exp_data.get_metadata()['sex'] == 'male'):
        male_ec_ids.append(exp["experiment_container_id"])
        print("EC ID " + str(i) + " added to male list.")
    else:
        female_ec_ids.append(exp["experiment_container_id"])
        print("EC ID " + str(i) + " added to female list.")
```
